### Back to the spherical collapse model

We can use equation for the collapse time, or redshift, predicted by the spherical collapse model (see <a href="l02_density_perturbations_evolution_tophat.ipynb">lecture 2 notebook</a>) to predict collapse times of shells of different radii $r$ that form a spherical perturbation with profile $\delta(<r)$. 

Assuming that $\delta_{i0}(<r_i)=\delta_i/D_+(t_i)$ is the initial overdensity profile linearly axtrapolated to $z=0$, $r_i$ is *comoving* radius of the initial shell, and growth function is normalized to one at $z=0$, i.e. $D_+(z=0)=1$, we can write:

$$\delta_{i0}(<r_i)D_+(z_{\rm coll})=\delta_{\rm c}$$

This is an implicit equation for $z_{\rm coll}$, which can be solved numerically for arbitrary cosmology for shells with different $r_i$. 

One $z_{\rm coll}$ is computed for a given shell $r_i$, the mass it encloses

$$M(<r_i)=\frac{4\pi}{3}\,\bar{\rho}_{\rm m0}\,[1+\delta(<r_i)]\,r_i^3 $$

has collapsed. Vectors of $z_{\rm coll}(r_i)$ (or $t_{\rm coll}$ - age of the universe at $z_{\rm coll}$) and $M(<r_i)$ for all shells comprise the mass collapse history of the peak. 

In analyses of collapse and evolution of dark matter halos in cosmological simulations, these are called *halo mass accretion history (MAH)*.

Time derivative of this function gives us *mass accretion rate* that such object experiences at any given time. 

$$\dot{M}_{\rm collapsed}=\frac{dM_{\rm collapsed}}{dz}\,\frac{dz}{dt}.$$

### For $\Omega_{\rm m}=1$, $\Omega_\Lambda=0$ cosmology solution is analytic: 

$D_+(z)=a(t)=(1+z)^{-1}$, so equation can be solved for $z_{\rm coll}$ analytically: 

$$z_{\rm coll} = \frac{\delta_i(<r_i)}{\delta_c} - 1$$

Note that even in the concordance $\Lambda$CDM model $\Omega_{\rm m}\approx 1$ at $z>2$ and the above equation provides a useful approximation for the redshift when a given radial shell will collapse. 

### $\Omega_{\rm m}=1$ solutions and scalings are applicable in $\Lambda$CDM at $z\geq 2-3$

### If we know initial $\delta(<r_i)$ profiles of peaks in the Gaussian $\delta(\mathbf{x})$ field (unsmoothed)

We could use such simple ansatz based on the spherical collapse model to predict the mass accretion history $M_{\rm collapsed}(z)$ for collapsed object that collapses from this peak. 

Such profiles can be computed for a Gaussian probability as averages over the appropriate conditional pdf, as described in some detail in the notebook (see <a href="http://adsabs.harvard.edu/abs/1986ApJ...304...15B">Bardeen et al. 1986</a>, <a href="http://adsabs.harvard.edu/abs/1987ApJ...318...15R">Ryden & Gunn 1987</a>, and <a href="http://adsabs.harvard.edu/abs/2010arXiv1010.2539D">Dalal et al. 2010</a> for additional details). 

### Bi-variate pdf and conditional pdfs

First, some background on conditional probabilities. If we have a bivariate pdf $p(x,y)$ of two random variables $x$ and $y$, we can use it to compute pdfs of single variables $x$ and $y$ as: 

$$p_x(x) = \int p(x,y)dy;\ \ \  p_y(y) = \int p(x,y)dy$$

as well as conditional pdf of $x$ given specific $y_\ast$ as 

$$p(x\vert y_\ast) = \frac{p(x,y)}{p_y(y_\ast)}.$$

The average of random variable $x$ over $p_x(x)$ is $\langle x\rangle = \int x p_x(x)dx.$ Similarly, average of $x$ under condition of specific value of $y_\ast$ can be computed as

$$\langle x\vert y_\ast\rangle = \int x p(x\vert y_\ast) dx.$$

### Bi-variate Gaussian conditional probabilities pdf and their mean and variance

When $p_{\rm G}(x,y)$ is bivariate Gaussian pdf, we have the following results for its averages and variance (see, e.g., <a href="http://athenasc.com/Bivariate-Normal.pdf">here</a>): 

$$\langle x\vert y_\ast \rangle = \frac{\langle xy\rangle}{\sigma^2_y}\, y_\ast +\langle x\rangle,$$ 

$$\sigma^2_{x\vert y_\ast} = \sigma^2_x  - \frac{\langle xy\,\rangle^2}{\sigma_y^2},$$ 

where 


$$\langle xy\rangle = \int xy\, p(x,y)dx dy; \ \ \ \sigma^2_y = \int y^2 p_y(y) dy.$$

### Multivariate Gaussian conditional probabilities with multiple simultaneous constraints

In the case of multivariate Gaussian, $p(x,\mathbf{y})$ where $\mathbf{y}$ is a vector of random variables, when we have multiple simultaneous constraints, instead of single $y_\ast$, the constraints can be arranged in a vector $\mathbf{y}_\ast$. The equations for the mean and variance above remain the same, but quantities such as $\langle x \mathbf{y}\rangle$ become vectors, while $\sigma_y^2=\langle \mathbf{y}\mathbf{y}\rangle$ becomes a matrix. 

### Initial density profiles of peaks in Gaussian peaks

For example, suppose we want to know the mean $\delta_r=\delta(<r)$ profiles around points in space that have specific value of $\delta_{R\ast}=\delta(<R)$, where $R$ is some specified radius we are interested in (corresponding to mass scale $M=4\pi/3\rho_{\rm m0} R^3$. 

This mean profile is given by the average $\langle\delta_r\vert\delta_{R\ast} \rangle$ as a function of $r$, where 
averaging is over the conditional Gaussian pdf $p(\delta_r\vert\delta_{R\ast})$: 

$$\bar{\delta}(<r\vert\delta_{R\ast})=\langle\delta_r\vert\delta_{R\ast} \rangle = \frac{\langle \delta_r\delta_R\rangle}{\langle\delta^2_R\rangle} \delta_{R\ast},$$

The variance around the mean $\langle\delta_r\vert\delta_{R\ast} \rangle$  is given by 

$$\sigma^2_{\delta_r\vert \delta_{R\ast}}=\langle \delta_r^2\rangle - \frac{\langle \delta_r\delta_R\rangle^2}{\langle\delta_R^2\rangle}.$$


In the equations above $\langle\delta^2_R\rangle$ and $\langle\delta^2_R\rangle$ are the usual variances of the Gaussian field smoothed on scale $R$ with tophat filter $W_{\rm TH}$:

$$\langle \delta_r^2\rangle=\frac{1}{2\pi^2}\int P(k)\vert W^2_{\rm TH}(k,r)\vert k^2 dk.$$
$$\langle \delta_R^2\rangle=\frac{1}{2\pi^2}\int P(k)\vert W^2_{\rm TH}(k,R)\vert k^2 dk.$$

and 

$$\langle\delta_r\delta_R\rangle = \frac{1}{2\pi^2}\int P(k)\vert W_{\rm TH}(k,r)W_{\rm TH}(k,R)\vert k^2 dk.$$


Profile $\langle\delta_r\vert\delta_R \rangle$ with the region of $\pm \sigma_{\delta_r\vert \delta_{R\ast}}$ for $M=1.92\times 10^{12}\, M_\odot$, or $R=1.69$ comoving Mpc in a $\Lambda$CDM cosmology with constraint $\delta_{R\ast}=2.73$ is shown in the plot below.

### Density profile taking into account existence of smaller $\delta$ sub-peaks

This can be done using a clever idea presented by <a href="http://arxiv.org/abs/1010.2539">Dalal et al. (2010)</a>.  

The idea is to compute an average value of the *maximum $\delta$* that $N=(R/r)^3$ subpeaks of radius $r$ can have within sphere of radius $R$, if the $\delta$ values of the subpeaks are randomly drawn from $p_{\rm G}(\delta_r \vert \delta_{R\ast})$.

 The main point to remember is that *unsmoothed density profile of a peak in a Gaussian field can be computed from the linear Gaussian field itself.* 

### Supremum statistics profile

If $x$ is a Gaussian random variable with pdf $p_{\rm G}(x)$ of zero mean and unit variance and we have $N$ samples $\{x_i\}$, the probability that a given $y$ is larger than any one $x_i$ is 
$$
P_1(y)=\int_{-\infty}^y\frac{dp_{\rm G}}{dx}dx = 1-\frac{1}{2}\mathrm{erfc}\left(\frac{y}{\sqrt{2}}\right),
$$

Assuming that $x_i$ samples are uncorrelated (not quite correct, but ignore correlations), the probability for $y$ to exceed all $N$ samples is: $P_{\rm N}(y) = [P_1(y)]^N.$ 

The differential pdf that $y$ is the supremum of $N$ samples of $x$ is $dP_N/dy$ and we can get an average value of $y$ in the usual way: 

$$\langle y\rangle = \int y\, \frac{dP_{\rm N}}{dy}\, dy = N\,\sqrt{\frac{2}{\pi}} \int y\,\left[1-\frac{1}{2}\mathrm{erfc}\left(y\right)\right]^{N-1} e^{-y^2}\, dy$$. 




### Supremum statistics profile

When we consider radius $r$ of the peak profile, we can think of $N\approx (R/r)^3$ sub-peaks of radius $r$ fitting within $R$. 

One of the sub-peaks has the largest $\delta$ and is located somewhere within sphere of radius $R$. We assume that it is this sub-peak that is identified as main progenitor in simulations. 

The sub-peaks have $\delta_r$ drawn from the Gaussian pdf $p_{\rm G}(\delta_r \vert \delta_{R\ast})$ for which we already know mean and variance. Thus, average *supremum* $\langle\delta_{r,\rm sup}\rangle$ can be computed using  $\langle y\rangle$ above as 

$$\langle\delta_{r,\rm sup}\rangle = \bar{\delta}(<r\vert\delta_{R\ast}) + \langle y\rangle\, \sigma_{\delta_r\vert \delta_{R\ast}} $$

### The takeaway points: 

* The mass accretion history $M(t)$ of collapsing dark matter halos *simply reflects the shape of the density profiles of the initial Gaussian peaks, $\delta(<r)$.*


* Each shell of the peak collapses at the time that is determined by its $\delta$ and collapse threshold that can be computed using simple spherical or ellipsoidal collapse model. 


### A useful approximation for average $\dot{M}(t)$

$$\dot{M}=  1.06\times 10^{12}\ M_\odot  \left(\frac{M}{10^{12}\ M_\odot}\right)^{1.14}\frac{\dot{D}_+(t)}{D^2_+(t)},
$$

where $D_+(t)$ is the linear growth factor for a given cosmology (<a href="https://ui.adsabs.harvard.edu/abs/2012ApJ...753...16K/abstract">Krumholz & Dekel 2012</a>). 



## Building a simple galaxy formation model

Starting at some initial time $t_i$ (redshift $z_i$) and collapsed seed mass $M_i$, let's assume mass is accreted onto evolving object at the rate $\dot{M}(t)$ that we can compute at any $t$. 

### Accretion rate of dark matter and baryons 

Let's assume that this mass comes as a mix of dark matter and baryonic diffuse gas, with baryons having a mass fraction $f_{\rm b,acc}$, i.e. mass accretion rate of baryons is 

$$\dot{M}_{\rm b,acc}= f_{\rm b,acc}\,\dot{M}=f_{\rm b,uni}\,f_{\rm b,in}\dot{M},
$$

where $f_{\rm b,uni}=\Omega_{\rm b}/\Omega_{\rm m}$ is "universal" baryon fraction (i.e., average over universe), while factor $f_{\rm b,in}$ allows for suppression of baryon accretion due to some physical process. In the simplest version we will assume that $f_{\rm b,in}=1$. 


### Evolution of gas mass 

Let's assume that all baryons are accreted as diffuse gas and that the accreted gas is instantly accreted onto galaxy in the center of evolving halo. This means that equation of the ISM gas mass evolution in a galaxy can be written as: 

$$
\dot{M}_{\rm g}=\dot{M}_{\rm b,acc}$$

In such evolution accreted gas would just accumulate in the galaxy. 

### Star formation

We know that galaxies convert accreted gas into stars via star formation process. For now let's just parametrize star formation rate on dimensional grounds as 

$$\dot{M}_\star =\frac{M_{\rm g}}{\tau_{\rm sf}}$$

where $\tau_{\rm sf}$ is called the *gas consumption timescale*. This ODE defines evolution of stellar mass. 

### Evolution of gas mass accounting for star formation

When stars form they return some fraction of their mass to the interstellar medium on $<10^8$ yrs time scale due to stellar winds blown by massive stars during their evolution, AGB stars shedding envelopes, and SN ejecta. 

This is short compared to Gyr time scales of galaxy evolution. So we can assume that the a fraction of stellar mass $\mathcal{R}$ is returned instantly (the *instantaneous recycling approximation.*): 

$$
\dot{M}_{\rm g}=\dot{M}_{\rm b,acc} - (1-\mathcal{R})\dot{M}_{\star}. $$


### Evolution of mass of heavy elements ("metals")  

In addition to return a fraction of mass to the ISM, young stars also synthesize and release new heavy elements (metals). The fraction of mass released in heavy elements is parameterized by the *yield*, $y_{\rm Z}\approx 0.07$ - fraction of mass of newly born stellar population that is returned to ISM in the form of metals. 

We can thus write equation for evolution of mass of heavy elements in the ISM: 

$$\dot{M}_{\rm Z}= y_{\rm Z}(1-\mathcal{R})\dot{M}_{\star}$$

### Evolution of mass of heavy elements ("metals")  

If we allow gas accreted from intergalactic medium to have some intrinsic *metallicity* $Z_{\rm IGM}$ then equation of evolution becomes: 

$$\dot{M}_{\rm Z}=Z_{\rm IGM}\dot{M}_{\rm g,acc} + y_{\rm Z}(1-\mathcal{R})\dot{M}_{\star}.$$


These equations can be numerically integrated as a system of ODEs. Such integration is implemented in <a href="codes/galaxy_model.py"><tt>codes/galaxy_model.py</tt></a> in python class <tt>model_galaxy</tt>. 

$$
\begin{cases}
\dot{M}_{\rm g, acc}= f_{\rm b,acc}\,\dot{M}=f_{\rm b,uni}\,f_{\rm b,in}\dot{M},\\
\dot{M}_\star =\frac{M_{\rm g}}{\tau_{\rm sf}} , \\
\dot{M}_{\rm g}=\dot{M}_{\rm b,acc} - (1-\mathcal{R})\dot{M}_{\star}, \\ 
\dot{M}_{\rm Z}=Z_{\rm IGM}\dot{M}_{\rm g,acc} + y_{\rm Z}(1-\mathcal{R})\dot{M}_{\star} 
\end{cases}$$


Code below shows how a galaxy model can be set up and integrated and how to plot results. Integration is done for $M_i=2.1\times 10^8\ M_\odot$ at $z_i=20$, with $\mathcal{R}=0.46$, $\tau_{\rm sf}=2$ Gyrs, and $Z_{\rm IGM}=10^{-4}$. 

```python
from colossus.cosmology import cosmology

# define a vector of cosmological parameters:    
my_cosmo = {'flat': True, 'H0': 70.0, 'Om0': 0.3, 'Ob0': 0.045714, 'sigma8': 0.81, 'ns': 0.96}
# set my_cosmo to be the current cosmology
cosmo = cosmology.setCosmology('my_cosmo', my_cosmo)    
```

```python
import codes.galaxy_model as gm
from scipy.integrate import odeint

def test_galaxy_evolution(Minit, sfrmodel, tausf, cosmo, verbose):
    zg = np.linspace(20., 0., 40) # grid of redshifts
    t_output = cosmo.age(zg) # and corresponding times

    # instantiate a galaxy model at t=tinit assuming a given star fromation model and cosmology
    g = gm.model_galaxy(t = t_output[0], Mh = Minit, Mg = None, Ms = None, MZ = None, 
                     sfrmodel = sfrmodel, tausf=tausf, cosmo = cosmo, verbose = verbose)
    
    # array of quantities to evolve with ODE solver
    y0 = np.array([g.Mh, g.Mg, g.Ms, g.MZ])
    # evolve y0 in time
    Mout = odeint(g.evolve, y0, t_output, rtol = 1.e-7, mxstep = 4000)
    
    # let's split the output into arrays with intuitive names
    Mhout = np.clip(Mout[:,0], 0.0, 1.e100)
    Mgout = np.clip(Mout[:,1], 0.0, 1.e100)
    Msout = np.clip(Mout[:,2], 0.0, 1.e100)
    MZout = np.clip(Mout[:,3], 0.0, 1.e100)

    return zg, t_output, Mhout, Mgout, Msout, MZout
```

```python
sfrmodel = "gaslinear" # star formation model; gaslinear is simply SFR=Mgas(t)/tau, where [tau]=[Gyrs] is constant 
tausf = 2. # gas consumption time scale in Gyrs
Zsun = 0.02 # solar metallicity

zg, t_output, Mhout, Mgout, Msout, MZout = test_galaxy_evolution(Minit=2.1e8, 
                                                                 sfrmodel=sfrmodel, 
                                                                 tausf=tausf, 
                                                                 cosmo=cosmo, 
                                                                 verbose=False)

```

```python
from scipy.interpolate import interp2d

def compute_mass_grid(gmodel = None, Nz=81, Nm=60, Z_IGM=2.e-5, 
                      sfrmodel='gaslinear', tausf=2.0,  cosmo=None):
    # grid of redshifts and halo masses
    zg = np.linspace(30., 0., Nz)
    lzg1 = np.log10(1.0+zg)
    t_output = cosmo.age(zg)
    
    lMh = np.linspace(5.0, 8., Nm)
    Mh  = np.power(10, lMh)

    # arrays to hold evolutionary grids of halo mass, gas mass, stellar mass and metal mass
    Mhout = np.zeros((Nm,Nz)); Mgout = np.zeros((Nm,Nz))
    Msout = np.zeros((Nm,Nz)); MZout = np.zeros((Nm,Nz))
    #evolve a grid of halo masses
    print("evolving mass grid...")
    for j, Mhd in enumerate(Mh):
        #if not j%5:
        #    print "evolving halo of log10(Mh)=%2.2f ..."%(lMh[j])
        # instantiate a model galaxy with a given halo mass at the initial time
        g = gmodel(t = t_output[0], Mh = Mhd, Mg = None, Ms = None, MZ = None,  
                       Z_IGM = Z_IGM, verbose = False, tausf = tausf, 
                       sfrmodel = sfrmodel,  cosmo = cosmo)
        # initial values of masses
        y0 = np.array([g.Mh, g.Mg, g.Ms, g.MZ])
        # solve the system of ODEs
        Mout = odeint(g.evolve, y0, t_output, rtol = 1.e-5, mxstep = 4000)
        # split Mout into arrays for specific masses with more intuitive names for convenience
        Mhout[j,:] = Mout[:,0]; Mgout[j,:] = Mout[:,1]
        Msout[j,:] = Mout[:,2]; MZout[j,:] = Mout[:,3]
    print("done.")

    # 
    # prepare 2D splines for interpolation
    #
    Mmin = 1.e-10 # prevent zeros in case no SF occurred
    lMhi = interp2d(lzg1, lMh, np.log10(np.maximum(Mmin,Mhout)), bounds_error=True, kind='cubic')
    lMgi = interp2d(lzg1, lMh, np.log10(np.maximum(Mmin,Mgout)), bounds_error=True, kind='cubic')
    lMsi = interp2d(lzg1, lMh, np.log10(np.maximum(Mmin,Msout)), bounds_error=True, kind='cubic')
    lMZi = interp2d(lzg1, lMh, np.log10(np.maximum(Mmin,MZout)), bounds_error=True, kind='cubic')

    return lMhi, lMgi, lMsi, lMZi
```

```python

lMhi, lMgi, lMsi, lMZi = compute_mass_grid(gmodel = gm.model_galaxy, 
                                                   Nz=41, Nm=60, Z_IGM=2.e-5, 
                                                   sfrmodel=sfrmodel, tausf=tausf, 
                                                   cosmo=cosmo)
```

### Re-ionization of the universe 

At $z_{\rm rec}\approx 1100$ the electrons and ions recombined, which allowed photons we know detect as the Cosmic Microwave Background (CMB) to propagate freely. 

Observations show, however, that by $z\approx 6$ intergalactic medium in the universe is ionized. So between $z=1100$ and $z\approx 6$ the gas must have been *re-ionized*. The process is called *reionization* and is almost certainly due to the UV photons emitted by young massive stars in the first galaxies at these high redshifts. 

### Re-ionization of the universe and heating of the intergalactic gas

The UV photons ionize neutral atoms and freed electrons carry extra energy which gets converted into gas temperature. Ionization by photons with spectrum typical of that of massive stars heats the gas to $T\approx 5000-10000 K$. This heating raises the *Jeans mass* of the integalactic gas and leads to suppression of gas accretion onto halos with the virial temperature:

$$kT_{\rm vir}=\mu m_p\frac{GM_{\rm vir}}{R_{\rm vir}}$$

less than the temperature of the gas. 

### Suppression of gas accretion on small-mass halos and "filtering mass" 

The mathematical formalism describing effect of such re-heating on the evolution of perturbations is presented in <a href="http://adsabs.harvard.edu/abs/2000ApJ...542..535G">Gnedin (2000).</a> The quantitative effect on accretion of baryons on a collapsing peak is calibrated using
 numerical simulations of galaxy formation  (e.g., <a href="http://adsabs.harvard.edu/abs/2006MNRAS.371..401H">Hoeft et al 2006</a>, <a href="http://adsabs.harvard.edu/abs/2008MNRAS.390..920O">Okamoto et al 2008</a>). It predicts that baryon fractions in halos of mass $M$ at redshift $z$ is modified from a constant value $\bar{f}_{\rm b}$ as:

$$
 f_{\rm b}(M,z)=\bar{f}_{\rm b}\left\{1+(2^{\alpha/3-1}-1)\left[\frac{M}{M_{\rm c}(z)}\right]^{-\alpha}\right\}^{-3/\alpha},
$$
 
with $\alpha=2$. 

The cutoff mass $M_{\rm c}$ is derived from simulations of galaxy formation.
For example, $M_{\rm c}(z)$, in figures of <a href="http://adsabs.harvard.edu/abs/2008MNRAS.390..920O">Okamoto et al (2008)</a>
is well approximated by the following formula:

$$
    M_{\rm c}(z) = 6\times 10^9h^{-1}\ M_\odot\,\exp(-0.63z),\ \ \mathrm{for}\ z<9,
$$

while $M_{\rm c}$ can be assumed to be $\approx 10^5-10^6h^{-1}\ M_\odot$ at $z>9$, because temperature of the gas is much smaller before reionization.  

Suppression of gas accretion is implemented using this formula in the variant of the galaxy formation model via $f_{\rm b,in}$ factor. 

$$
 f_{\rm b}(M,z)=\bar{f}_{\rm b}\left\{1+(2^{\alpha/3-1}-1)\left[\frac{M}{M_{\rm c}(z)}\right]^{-\alpha}\right\}^{-3/\alpha},
$$

### Code to compute average profile of a Gaussian peak smoothed on a given scale. 

Calculation is based on the equations above (derived in <a href="http://adsabs.harvard.edu/abs/1986ApJ...304...15B">Bardeen et al. 1986</a>, <a href="http://adsabs.harvard.edu/abs/1987ApJ...318...15R">Ryden & Gunn 1987</a>, and <a href="http://adsabs.harvard.edu/abs/2010arXiv1010.2539D">Dalal et al. 2010</a>).

Here for comparison with simulation results a particular cosmological model of that simulation is adopted. 

```python
from codes.variances import delta_variance, dvars, ddvars, ex_mean
from colossus.halo.mass_so import M_to_R
from colossus.lss import peaks

def compute_ave_dpro_for_deltaR(cosmo, MR, dRstar, ri, expro=False, verbose=False):
    rtol = 1.e-8

    RTH = peaks.lagrangianR(MR*cosmo.h) / (1. + dRstar)**(1./3.)
    if verbose: 
        zd = 0.
        print("R200m = %.3f kpc at z = %.3f"%(M_to_R(MR*cosmo.h, zd, '200m')/cosmo.h, zd))
        print("R_TH = %.3f comoving Mpc"%(RTH/cosmo.h))

    dRR = delta_variance(RTH, RTH, func=dvars, rtol=rtol, cosmo=cosmo)

    dlproave = np.zeros_like(ri)
    vardlproave = np.zeros_like(ri)

    for ir, rd in enumerate(ri):
        # <delta_r^2>
        var   = delta_variance(rd, rd, func=dvars, rtol=rtol, cosmo=cosmo)
        # <delta_r x delta_R>
        vardd = delta_variance(rd, RTH, func=ddvars, rtol=rtol, cosmo=cosmo)
        # average delta(<r) profile using equation above 
        dlproave[ir] = vardd / dRR * dRstar
        # variance about the average profile, can't be <0
        vardlproave[ir] = np.maximum(0., var - vardd**2 / dRR)
        
    if expro:
        for ir, rd in enumerate(ri):
            Nr = (RTH/rd)**3
            xex = 0.
            if Nr > 1:
                xex = ex_mean(Nr, rtol=rtol)
                dlproave[ir] = xex*np.sqrt(vardlproave[ir]) + dlproave[ir]

    return dlproave, vardlproave
```

```python
from scipy.interpolate import UnivariateSpline

def dpro_to_mah(cosmo, zmin, zmax, nzg, ri, dproi, delta_c=None):
    zg = np.linspace(zmax, zmin, nzg)
    # corresponding age of the universe at zg
    tg = cosmo.age(zg)
    #overdensity required to collapse by zg
    Dgr = cosmo.growthFactor(zg)
    # interpolate computed profile with cubic spline, so we can easily compute its derivative
    lDzsp  = UnivariateSpline(np.log10(Dgr), np.log10(1.+zg), s=0.)

    Mri = 4 * np.pi / 3. * (1. + dproi) * cosmo.rho_m(z=0.) * ri**3 * 1.e9 / cosmo.h
    zcr = np.maximum(10.**lDzsp(np.log10(delta_c/dproi)) - 1., -0.998)
    return zcr, Mri
```

### Fixing low-$z$ MAH by constraining slope at $r=R$ to be steep

Physically, such constraint ensures that peak is relatively isolated and is not located next to another large overdensity. 

We can compute average density profile of Gaussian peaks as follows  (<a href="http://adsabs.harvard.edu/abs/1986ApJ...304...15B">Bardeen et al. 1986</a>, <a href="http://adsabs.harvard.edu/abs/1987ApJ...318...15R">Ryden & Gunn 1987</a>, and <a href="http://adsabs.harvard.edu/abs/2010arXiv1010.2539D">Dalal et al. 2010</a>).

If we consider a peak of mass $M$ and radius $R$ and overdensity $\delta(<R)\equiv\delta_R$ within $R$, i.e. $M=(4\pi/3)(1+\delta_R)\bar{\rho}R^3$. Let's also define derivative of $\delta_R$ at $R$:

$$\delta^\prime_R=\frac{d\delta_R}{dr}.$$

Then, average profile $\bar{\delta}(<r)\equiv\bar{\delta}_r$ of peaks is given by the average over the conditional pdf of profiles $\delta_r$ (see, e.g., p.3-4 <a href="http://athenasc.com/Bivariate-Normal.pdf">here</a> for mean and variance of conditional Gaussian pdf) of individual peaks:

$$\bar{\delta}(<r)=\langle p(\delta_r\vert \delta_R, \delta^\prime_R)\rangle=\langle X\vert Y\rangle = \langle XY\rangle\langle YY\rangle^{-1} Y,$$

where $X = \delta_r$ and $Y=[\delta_R,\delta^\prime_R]$. The variance of the profile is given by 

$$\sigma^2_{X\vert Y}=\langle XX\rangle - \langle XY\rangle\langle YY\rangle^{-1} \langle YX\rangle.$$

Here, specifically: 

$$\langle XX\rangle = \sigma^2(r)=\frac{1}{2\pi^2}\int P(k)\vert W_{\rm TH}(k,r)\vert^2 k^2 dk,\ \ \ W_{\rm TH}=\frac{3(\sin x - x\cos x)}{x^3};\ \ x\equiv kr.$$

$$\langle YX\rangle = \langle XY\rangle^T = \left[\langle\delta_r\delta_R\rangle,\langle\delta_r,\delta_R^\prime\rangle\right].$$

$$\langle\delta_r\delta_R\rangle = \frac{1}{2\pi^2}\int P(k)\vert W_{\rm TH}(k,r)W_{\rm TH}(k,R)\vert k^2 dk.$$

$$\langle\delta_r\delta^\prime_R\rangle = \frac{1}{2\pi^2}\int P(k)\vert W_{\rm TH}(k,r)\,\frac{\partial W_{\rm TH}}{\partial R}(k,R)\vert k^2 dk.$$

$$\langle YY\rangle=\left[\begin{matrix} 
\langle\delta_R^2\rangle & \langle\delta_R\delta_R^\prime\rangle \\
\langle\delta_R\delta^\prime_R\rangle& \langle\delta^{\prime 2}_R\rangle 
\end{matrix}\right]$$

```python
from codes.variances import delta_variance, dvars, ddvars, dddvars, ddddvars
from scipy.special import erfc
from codes.variances import ex_mean

from colossus.halo.mass_so import M_to_R
from colossus.lss import peaks

def compute_ave_dpro_for_deltaR_slope(cosmo, MR, dRstar, dddRstar, ri, expro=False, verbose=False):
    rtol = 1.e-9

    RTH = peaks.lagrangianR(MR*cosmo.h) / (1. + dRstar)**(1./3.)
    if verbose: 
        zd = 0.
        print("R200m = %.3f kpc at z = %.3f"%(M_to_R(MR*cosmo.h, zd, '200m')/cosmo.h, zd))
        print("R_TH = %.3f comoving Mpc"%(RTH/cosmo.h))

    Y = np.zeros(2); 
    Y[0] = dRstar; Y[1] = dddRstar # -6.5 for Via Lactea 2 specifically

    YY = np.zeros((2,2))
    YY[0,0] = delta_variance(RTH, RTH, func=dvars, rtol=rtol, cosmo=cosmo)
    YY[0,1] = delta_variance(RTH, RTH, func=dddvars, rtol=rtol, cosmo=cosmo)
    YY[1,0] = YY[0,1]
    YY[1,1] = delta_variance(RTH, RTH, func=ddddvars, rtol=rtol, cosmo=cosmo) 
    YYi = np.linalg.inv(YY)

    XY = np.zeros(2); 
    dlproave = np.zeros_like(ri)
    vardlproave = np.zeros_like(ri)

    for ir, rd in enumerate(ri):
        var = delta_variance(rd, rd, func=dvars, rtol=rtol, cosmo=cosmo)
        vardd = delta_variance(rd, RTH, func=ddvars, rtol=rtol, cosmo=cosmo)
        varddd = delta_variance(rd, RTH, func=dddvars, rtol=rtol, cosmo=cosmo)
    
        XY[0] = vardd; XY[1] = varddd
        dlproave[ir] = np.dot(XY,np.dot(YYi,Y))
        vardlproave[ir] = np.maximum(0., var - np.dot(XY,np.dot(YYi,XY.T)))

    if expro:
        for ir, rd in enumerate(ri):
            Nr = (RTH/rd)**3
            xex = 0.
            if Nr > 1:
                xex = ex_mean(Nr, rtol=rtol)
                dlproave[ir] = xex*np.sqrt(vardlproave[ir]) + dlproave[ir]

    return dlproave, vardlproave
```
